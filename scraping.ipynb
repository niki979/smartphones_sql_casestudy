```python
import requests
from bs4 import BeautifulSoup
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.common.exceptions import TimeoutException
from selenium.webdriver.edge.service import Service
import time
import pandas as pd
```

```python
edge_service = Service('C:/Users/Dell/Downloads/edgedriver_win64/msedgedriver.exe')

# Start the service
edge_service.start()

# Create the Microsoft Edge webdriver
driver = webdriver.Edge(service=edge_service)
```

```python
driver.get('https://www.smartprix.com/mobiles')

old_height = driver.execute_script('return document.body.scrollHeight')

while True:
    button_element = driver.find_element(By.XPATH, '//*[@id="app"]/main/div[1]/div[3]/div[3]')
    button_element.click()
    time.sleep(3)

    
    new_height = driver.execute_script('return document.body.scrollHeight')
        
    # Check if the page height remains unchanged, indicating no more content to load
    if new_height == old_height:
        break
    old_height = new_height
```

```python
soup = BeautifulSoup(driver.page_source, 'html.parser')
```

```python
name=[]
price=[]
rating=[]
score=[]
sim=[]
processor=[]
ram=[]
battery=[]
display=[]
camera=[]
card=[]
os=[]
```

```python
table=soup.find('div',class_='sm-products list size-m img-long len-1020')
mobiles=table.findAll('div',class_='sm-product has-tag has-features has-actions')
for i in mobiles:
    name.append(i.find('h2').text)
    price.append(i.find('span',class_='price').text)
    score.append(i.find('b').text)
    ls=i.find('ul',class_='sm-feat specs').find_all('li')
    try:
        sim.append(ls[0].text)
    except:
        sim.append('')
    try:
        processor.append(ls[1].text)
    except:
        processor.append('')
    try:
        ram.append(ls[2].text)
    except:
        ram.append('')
    try:
        battery.append(ls[3].text)
    except:
        battery.append('')
    try:
        display.append(ls[4].text)
    except:
        display.append('')
    try:
        camera.append(ls[5].text)
    except:
        camera.append('')
    try:
        card.append(ls[6].text)
    except:
        card.append('')
    try:
        os.append(ls[7].text)
    except:
        os.append('')

    
```

```python
df=pd.DataFrame(
{
    'name':name,
    'price':price,
    'score':score,
    'sim':sim,
    'processor':processor,
    'ram':ram,
    'battery':battery,
    'display':display,
    'camera':camera,
    'card':card,
    'os':os

}
)
```
